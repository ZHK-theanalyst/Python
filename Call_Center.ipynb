```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel(r"C:\Users\USER\Documents\Data Analysis\PandasYouTubeSeries-main\Customer Call List.xlsx")
df1
```

```python
df1 = df1.drop_duplicates()

df1
```

```python
df1 = df1.drop(columns = 'Not_Useful_Column')
df1
```

```python
df1['Last_Name'] = df1['Last_Name'].str.strip('123._/')

df1
```

```python
df1
```

```python
df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].astype(str).str.replace(r'[-|/]','', regex = True)

df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].apply(lambda x: f"{x[:3]}-{x[3:6]}-{x[6:10]}" if len(x) == 10 else x)

df1
```

```python
df1
```

```python
df1["Phone_Number"] = df1["Phone_Number"].str.replace('nan','')
df1["Phone_Number"] = df1["Phone_Number"].str.replace('Na','')
df1
```

```python
df1[['Street_Address','State','Zip_Code']] = df1['Address'].str.split(',',n=2,expand = True)

df1
```

```python
df1['Paying Customer'] = df1['Paying Customer'].str.replace('Yeseses', 'Yes').str.strip().str.lower().map({'yes': 'Yes', 'y': 'Yes', 'no': 'No', 'n': 'No'})
df1
```

```python
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].str.replace('Y','Yes').str.strip().str.lower().map({'yes' : 'Yes', 'y' :'Yes', 'no' : 'No', 'n' :'No'})

df1
```

```python
df1['Address'] = df1['Address'].str.replace('N/a','')

df1
```

```python
df1['Paying Customer'] = df1['Paying Customer'].fillna('')
df1['Do_Not_Contact'] = df1['Do_Not_Contact'].fillna('')

df1
```

```python
df1 = df1.fillna('')

df1
```

```python
df1
```

```python
df_filtered = df1.dropna(subset = ['Phone_Number', 'Do_Not_Contact'])
df_filtered= df_filtered[df_filtered["Do_Not_Contact"] =='No']

df_filtered
```

```python
for x in df_filtered.index:
    if df_filtered.loc[x, 'Phone_Number'] == '':
        df_filtered.drop(x, inplace = True)

df_filtered
```

```python
df_filtered = df_filtered.drop(columns = 'Address')

df_filtered
```
